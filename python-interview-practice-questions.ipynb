## 8. Question: Write a function that returns the n-th Fibonacci number using recursion.

```python
def fibbonacci_num(n):
    if n <= 1:
        return n
    else:
        # Recursive case: sum of the two preceding numbers
        return fibbonacci_num(n-1)+fibbonacci_num(n-2)
    
# Example Usage
n = 6
print(f"The {n}th fibbonacci number is {fibbonacci_num(n)}")
    
```

### Explanation:

    Base cases:
        fibonacci(0) returns 0
        fibonacci(1) returns 1
    Recursive case: For any n > 1, the function recursively calls itself to compute fibonacci(n-1) + fibonacci(n-2).

Example calculation for n = 6:

    fibonacci(6) = fibonacci(5) + fibonacci(4)
    fibonacci(5) = fibonacci(4) + fibonacci(3)
    fibonacci(4) = fibonacci(3) + fibonacci(2)
    fibonacci(3) = fibonacci(2) + fibonacci(1)
    fibonacci(2) = fibonacci(1) + fibonacci(0)

Finally, this results in fibonacci(6) = 8.
Note:

Recursive solutions to the Fibonacci sequence are easy to understand but not the most efficient, as they involve recalculating values multiple times. For larger n, a more efficient approach like dynamic programming (memoization) would be better.

A decorator is like a wrapper around a function. It lets you add new functionality to the function without changing the function itself. You can imagine it as a gift wrap: the gift is still the same inside, but the decorator (wrapper) changes how the gift looks when you give it to someone.


```python
def my_decorator(func):
    def wrapper():
        print("Print something before the function")
        func() # This is the original function call
        print("Print something after the function")
    return wrapper()

@ my_decorator
def say_hello():
    print("Hello")
```

## 13. Question: Explain the concept of *args and **kwargs in Python.
### Answer:
*args allows you to pass a variable number of positional arguments to a function, while **kwargs allows you to pass a variable number of keyword arguments.

#### 1. *args (Positional Arguments):

```python
def my_arg_func(*args):
    for arg in args:
        print(arg)
        
my_arg_func(1,2,3,4,5)
        
```

#### Explanation:

    The *args syntax allows the function to accept any number of positional arguments.
    The arguments passed to the function are collected into a tuple args inside the function.

#### 2. **kwargs (Keyword Arguments):

```python
def my_kwargs_func(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")
              
my_kwargs_func(name="Ern", age=30, location="Sri Lanka")
```

#### Explanation:

    The **kwargs syntax collects keyword arguments into a dictionary kwargs.
    This allows the function to handle a variable number of keyword arguments dynamically.

#### 3. Combining *args and **kwargs:

You can use both *args and **kwargs in the same function to accept both positional and keyword arguments.

```python
def my_func(*args, **kwargs):
    print("Possitional arguments: ", args)
    print("Keyward arguments: ", kwargs)
    
my_func(1,2,3, name="bob", age=28)
```

#### Explanation:

    *args collects the positional arguments (1, 2, 3).
    **kwargs collects the keyword arguments {'name': 'Alice', 'age': 30}.

#### 4. Why use *args and **kwargs?

    Flexibility: Functions can accept any number of arguments without pre-defining them.
    Dynamic code: Allows for more dynamic function calls, especially useful in situations where you don't know the number of arguments in advance.

Summary:

    *args is used to pass a variable number of positional arguments as a tuple.
    **kwargs is used to pass a variable number of keyword arguments as a dictionary.
    You can use both in the same function to handle both types of arguments simultaneously.

## 15. Question: How can you achieve multi-threading in Python?
### Answer:
Python has a threading module which can be used to achieve multi-threading.
In Python, multi-threading can be achieved using the threading module, which allows you to run multiple threads (smaller units of a process) concurrently. Each thread runs independently and can perform tasks in parallel, which can be useful for I/O-bound tasks like reading from files, handling network requests, or performing tasks where waiting (e.g., for input/output) happens.

However, due to Python’s Global Interpreter Lock (GIL), multi-threading does not give a true parallel execution of threads for CPU-bound tasks (tasks that require heavy computation). Instead, threads take turns executing. For CPU-bound tasks, using the multiprocessing module is generally more effective.
Steps to Achieve Multi-threading:

    Import the threading module.
    Define a function that represents the task each thread will perform.
    Create Thread objects and pass the target function to them.
    Start the threads using the start() method.
    Wait for the threads to complete using join().

```python
import threading

# Function to be executed by each thread
def print_numbers():
    for i in range(1,6):
        print(i)

def print_letters():
    for letter in 'abcde':
        print(letter)
        
# Creating multiple threads
t1 =threading.Thread(target=print_numbers)
t2 = threading.Thread(target=print_letters)

# Starting the threads
t1.start()
t2.start()

# Waiting for both threads to complete
t1.join()
t2.join()

print("Both threads has completed the execution.")
```

## 21. Question: How can you implement a singleton pattern in Python?

### Answer:
One way is by using a class attribute to check if an instance already exists. If it does, return that. Otherwise, create a new instance.
Key Characteristics of a Singleton

    Single Instance: It ensures only one instance of the class is created.
    Global Access: Provides a way to access this instance globally.

```python
class Singleton:
    _instance = None # Private attribute to store the single instance of the class
    
    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance
    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance
    
# Example usage
singleton1 = Singleton()
singleton2 = Singleton()
print(singleton1 is singleton2) # Output: True (both are the same instance)
```

## 24. Question: Explain Python's garbage collection process.
### Answer:
Python's garbage collection is a process that automatically manages memory, reclaiming it when objects are no longer needed to prevent memory leaks. Python primarily uses two mechanisms for garbage collection: reference counting and cyclic garbage collection.

### 1. Reference Counting:
        Each object in Python has a reference count that tracks how many references point to it.
        When an object’s reference count drops to zero (meaning no variable or object references it anymore), Python immediately frees up the memory.
        Although simple and efficient, reference counting has a limitation—it cannot detect circular references (where two or more objects reference each other).


```python
a = []
b = a
del a
del b # Memory for the empty list is freed because no references exist
```

  ### 2. Cyclic Garbage Collection:
        To handle circular references, Python has a cyclic garbage collector as part of its gc module.
        This collector periodically looks for groups of objects that reference each other (cycles) and are unreachable from outside.
        It runs in generations, grouping objects by age, and focuses on collecting newer objects more frequently, as these are more likely to be eligible for garbage collection.

Managing Garbage Collection:

    Python’s garbage collection can be manually managed using the gc module, which allows enabling, disabling, and adjusting garbage collection thresholds.
    This is typically not needed, as Python’s garbage collection is automatic and efficient, but it can be useful in memory-intensive programs.

```python
import gc
gc.collect() # Manually trigger garbage collection/ returns the number of unreachable objects it has collected
```

Summary:

    Reference Counting: Immediate collection for objects with zero references.
    Cyclic Garbage Collection: Periodic cleanup of circular references.

## 33. Question: What are context managers in Python? Provide an example.
### Answer:
In Python, context managers are constructs that allow you to manage resources efficiently, ensuring that setup and cleanup tasks are handled automatically. They are typically used when working with resources like files, network connections, or database connections, where proper cleanup (e.g., closing a file) is essential.

Context managers are commonly used with the with statement, which guarantees that the resource is correctly closed or released when the block of code is done, even if an error occurs.
Example of a Context Manager with Files

One of the most common uses of a context manager is managing file resources. When you open a file using with, it ensures that the file is closed after the block is exited, no matter how it’s exited.


```python
# Using a conext manager to open and read files
with open('example.txt', 'r') as file:
    contents = file.bufferread()
    print(contents)
# No need to call file.close() – it's automatically done after the block
```

### In this example:

    open("example.txt", "r") opens the file in read mode.
    with ensures that the file is closed automatically when the block ends, even if an error occurs inside the block.

Creating a Custom Context Manager

You can create your own context manager by defining a class with __enter__() and __exit__() methods. The __enter__() method is run when entering the with block, and __exit__() is run when exiting the block.

```python
# Custom context manger to handle a simple resources
class MyContextManager:
    def __enter__(self):
        print("Entering the context..")
        return "Resource is ready..."
    
    def __exit__(self, exc_type, exc_value, traceback):
        print("Existing the context..")
        # Cleanup code if any
        if exc_type:
            print(f'An error occured {exc_value}')
        return True # Handle exception if neccessary
    
# Using the custom context manager
with MyContextManager() as resource:
    print(resource)
```

### Example Using contextlib to Create a Context Manager

Python’s contextlib module provides an easier way to create context managers using a generator function and the @contextmanager decorator.

```python
from contextlib import contextmanager

@contextmanager
def My_context():
    print("Setup")
    yield "Resource is ready.."
    print("Cleanup")
    
with My_context() as resource:
    print(resource)
```

### Summary

    Context managers ensure that resources are properly managed, handling setup and cleanup tasks automatically.
    They are most commonly used with the with statement, ensuring resources are released, even if errors occur.
    You can create a custom context manager by defining __enter__() and __exit__() in a class or by using @contextmanager with a generator.

## 5. Question: What does the zip function do in Python?

### Answer:
The zip function in Python is used to combine multiple iterables (like lists, tuples, or strings) into a single iterable of tuples. Each tuple contains elements from the corresponding positions of the input iterables. If the input iterables are of different lengths, zip stops creating tuples when the shortest input iterable is exhausted.

### Syntax
zip(*iterables)

```python
# Two list
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]

# Combine two lists
result = zip(names, ages)

# convert into list to see the result
print(list(result))
```

### Explanation

    In this example, zip takes two lists: names and ages.
    It combines them into a list of tuples, where each tuple contains one name and the corresponding age.

Handling Different Lengths

If the input iterables are of different lengths, zip will stop when the shortest iterable is exhausted:

```python
names = ['Alice', 'Bob']
ages = [20, 30, 35]

result = zip(names, ages)

print(list(result))
```

### In this case, the output contains only the first two pairs because the ages list has three elements, but the names list has only two. Thus, zip stops after the shortest iterable (names) is exhausted.
Unzipping

You can also "unzip" a zipped list back into individual lists:

```python
names = ['Alice', 'Bob']
ages = [20, 30, 35]

combined = list(zip(names, ages))

print(combined)

# Unzipping the zipped list
if combined:
    unzipped_names, unzipped_ages = zip(*combined)
    print(unzipped_names)
    print(unzipped_ages)
```

This is useful for scenarios where you want to separate combined data back into its original form.
