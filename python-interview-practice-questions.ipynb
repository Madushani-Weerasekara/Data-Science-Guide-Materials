1. Question: Write a function that takes a list of numbers and returns the sum.

```python
def sum_of_list_num(num):
    return sum(num)

# Example Usage
num_list = [1,2,3,4,5]
sum_of_list_num(num_list)
```

2. Question: How do you reverse a string in Python?

```python
def rev_string(s):
    return s[::-1]

# Example uSAGE
my_string = "Madushani"

rev_string(my_string)
```

4. Question: Write a function that checks if a given word is a palindrome.

```python
def is_palindrome(word):
    return word == word[::-1]

# Example Usage
word = "maam"
is_palindrome(word)
    
```

5. Question: How can you remove duplicates from a list?

You can remove duplicates from a list in Python by converting the list to a set, which automatically removes duplicates, and then converting it back to a list.

```python
def remove_duplicates(lst):
    return list(set(lst))

# Example Usage
my_list = [8,4,4,2,6,8,10]
remove_duplicates(my_list)
```

```python
def remove_duplicates(lst):
    result = []
    for item in lst:
        if item not in result:
            result.append(item)
    return result

# Example Usage
my_list = [8,4,4,2,6,8,10]
remove_duplicates(my_list)
```
