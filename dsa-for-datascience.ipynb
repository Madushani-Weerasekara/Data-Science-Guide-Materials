# Lets say two dictionaries food and orders by considering the total time for each order sort them in accending order.

```python
food = {
    "Koththu" : 10,
    "CRice" : 9,
    "CheeseKoththu" : 20,
    "Milkshake" : 5,
    "Chips" : 2,
    "Desert" : 6,
    "Parata" : 3,
    "BeefKoththu" : 25,
    "Biriyani" : 30,
    "Roti": 5,
    "BeefCurry" : 22,
    "ChickenCurry" : 21,
    "MuttonCurry" : 40
            }

food
```

```python
import random

orders = {}

# List of available food items for generating random combinations
food_items = list(food.keys())

for i in range(1, 21): # loop to create 20 orders
    order_items = {}

    # Randomly select 2-5 food items per order
    items_in_order = random.sample(food_items, random.randint(2, 5))

    for item in items_in_order:
        # Random quantity between 1 and 3 for each item in the order
        order_items[item] = random.randint(1, 3)

    # Add the generate order to the main orders dictionary
    orders[str(i)] = order_items
print(orders, sep="\n")
    
```
