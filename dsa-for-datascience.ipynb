# Lets say two dictionaries food and orders by considering the total time for each order sort them in accending order.

```python
food = {
    "Koththu" : 10,
    "CRice" : 9,
    "CheeseKoththu" : 20,
    "Milkshake" : 5,
    "Chips" : 2,
    "Desert" : 6,
    "Parata" : 3,
    "BeefKoththu" : 25,
    "Biriyani" : 30,
    "Roti": 5,
    "BeefCurry" : 22,
    "ChickenCurry" : 21,
    "MuttonCurry" : 40
            }

food
```

```python
import random

orders = {}

# List of available food items for generating random combinations
food_items = list(food.keys())

for i in range(1, 21): # loop to create 20 orders
    order_items = {}

    # Randomly select 2-5 food items per order
    items_in_order = random.sample(food_items, random.randint(2, 5))

    for item in items_in_order:
        # Random quantity between 1 and 3 for each item in the order
        order_items[item] = random.randint(1, 3)

    # Add the generate order to the main orders dictionary
    orders[str(i)] = order_items
print(orders, sep="\n")
    
```

```python
orders
```

```python
def calculate_time(food, orders):
    
    resdict = {}

    # aculate the preparation time for each order
    for orderId, items in orders.items(): # Outer loop : looping through the orders 
        
        orderTime = 0 # orderTime begin of each order is set to zero

        for item, quantity in items.items(): # Inner loop
            if item in food:
                orderTime += food[item] * quantity
        
        resdict[orderId] = orderTime # store the total time for the current order

    # sort the orers by their preparation time
    sorted_dict = dict(sorted(resdict.items(), key=lambda x : x[1]))

    # Print the sorted orders
    for orderId, time in sorted_dict.items():
        print(f'Order {orderId} : {time} minutes')

calculate_time(food,orders)


 
```
