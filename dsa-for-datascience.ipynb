```python
import random

orders = {}

# List of available food items for generating random combinations
food_items = list(food.keys())

for i in range(1, 21): # loop to create 20 orders
    order_items = {}

    # Randomly select 2-5 food items per order
    items_in_order = random.sample(food_items, random.randint(2, 5))

    for item in items_in_order:
        # Random quantity between 1 and 3 for each item in the order
        order_items[item] = random.randint(1, 3)

    # Add the generate order to the main orders dictionary
    orders[str(i)] = order_items
print(orders, sep="\n")
    
```

```python
def calculate_time(food, orders):
    
    resdict = {}

    # aculate the preparation time for each order
    for orderId, items in orders.items(): # Outer loop : looping through the orders 
        
        orderTime = 0 # orderTime begin of each order is set to zero

        for item, quantity in items.items(): # Inner loop
            if item in food:
                orderTime += food[item] * quantity
        
        resdict[orderId] = orderTime # store the total time for the current order

    # sort the orers by their preparation time
    sorted_dict = dict(sorted(resdict.items(), key=lambda x : x[1]))

    # Print the sorted orders
    for orderId, time in sorted_dict.items():
        print(f'Order {orderId} : {time} minutes')

calculate_time(food,orders)


 
```

# Traffic Light System Simulation

## Scenario:
Imagine designing a traffic control system for a busy intersection. Cars should wait at red lights, and you need to minimize waiting time by efficiently managing the order of signals.

## Questions:
        How would you use data structures to manage the queue of cars at each traffic light?
        What algorithm would you implement to minimize waiting time for cars?
        How would you handle emergencies (e.g., ambulances) to bypass normal queues?
        
DSA Concepts: Queues (for each traffic light), priority queues (for emergency vehicles), and greedy algorithms (for minimizing wait times).

### 1. Using Data Structures to Manage the Queue of Cars at Each Traffic Light

### Solution:
Use queues to manage cars waiting at each light, as queues efficiently handle the "first-in, first-out" (FIFO) order that traffic lights follow.

Each direction at the intersection (e.g., north, south, east, west) can have its own queue, representing cars waiting at that particular signal.

These queues can be represented as arrays, linked lists, or, in a more sophisticated implementation, as a standard queue data structure in programming languages like Python, Java, or C++.

##  Algorithm to Minimize Waiting Time for Cars

### Solution:
Use a greedy algorithm that continuously evaluates which queue has the longest wait time or the highest number of cars and gives it priority

Implement a dynamic timing mechanism that adjusts signal duration based on the length of each queue. This approach is also called adaptive traffic control.

For example, if the east queue is much longer than others, the green light for the eastbound direction might last longer or turn green sooner.

### Algorithm Steps:
Periodically assess the length of each queue (e.g., every few seconds).

Prioritize the queue with the highest number of cars.

Adjust the duration of the green light based on the queue length, allowing more cars to pass if a queue is long.

Rotate to the next longest queue to keep traffic balanced.

## 3. Handling Emergencies (e.g., Ambulances) to Bypass Normal Queues

### Solution:

Use a priority queue or heap data structure to prioritize emergency vehicles.

If an emergency vehicle (like an ambulance or fire truck) approaches, it is given higher priority over regular vehicles.

The system should detect this and temporarily override the queue sequence to give the emergency vehicle an immediate green light.

Priority queues allow us to prioritize based on urgency. For instance, emergency vehicles can be tagged with the highest priority value.

### mplementation Example:

Each queue can be enhanced to hold both regular cars and emergency vehicles, using a priority flag.

Emergency vehicles are enqueued with high priority, and as soon as they’re detected, the system checks if there’s an emergency in any queue. If so, the system grants green light access to that direction immediately.

```python
# deque data structure to manage the queue of cars

from collections import deque
from time import sleep
import heapq # Heap data structure to handle mergencies to by pass normal queues

north_queue = deque(['car1', 'car2', 'car3']) # queue for the cars waiting at the north signal
south_queue = deque(['car1', 'car2', 'car3', 'car4', 'car5']) # queue for the cars waiting at the south signal
east_queue = deque(['car1', 'car2', 'car3','car4', 'car5', 'car6', 'car7', 'car8', 'car9', 'car10']) # queue for the cars waiting at the east signal
west_queue = deque(['car1', 'car2', 'car3','car4', 'car5','car6', 'car7']) # queue for the cars waiting at the west signal

# Greedy Algorithm to minimize cars waiting time

queues = {'north' :  north_queue, 'south' : south_queue, 'east' : east_queue, 'west' : west_queue}

#priority queue for handling emergencies
emergency_queue = []

# Display each queue and its length for each direction
for direction, queue in queues.items():
    print(f'Direction : {direction} Queue : {list(queue)} Length : {len(queue)}')
    print() # new line

def green_light(direction):
    """
    Simulates a green light for a specific direction.
    Processes cars in the queue for that direction.
    """
    print(f"The light is green for {direction.upper()}. Cars are passing...")

     # Allow cars to pass one by one
    while queues[direction]:
        car = queues[direction].popleft() # remove the first car in the queue
        print(f'Car {car} has passed..')
        sleep(1)  # Simulate time for a car to pass
    print(f'No more cars waiting at {direction.upper()}.\n')
    

def choose_next_light():
    """
    Determines which light should turn green next based on the longest queue.
    """
    # find the direction with longest queue
    max_queue = max(queues, key=lambda x : len(queues[x]))
    print("Next light to turn green :", max_queue)
    return max_queue
    


def handle_emergencies(direction):
     """
    Handles emergency vehicles by granting them immediate access.
    """
    print(f'Emergency vehicle detected in {direction.upper()} direction!')
    # Insert emergency vehicle with highest priority (e.g., -1 priority)
    heapq.heappush(emergency_queue, (-1, direction))

def check_emergencies():
    """
    Checks if there are any emergencies and handles them.
    """
    if emergency_queue:
        print(f'Handling emergency in {direction.upper()} direction!')
        # Insert emergency vehicle with highest priority (e.g., -1 priority)
        priority, direction = heapq.heappop(emergency_queue)
        green_light(direction)

        # Grant emergency vehicle immediate access
        sleep(5) # Short green duration to allow emergency vehicle to pass


def adaptive_signal_control():
    while any(len(queue)) > 0 for queue in queues.valuers()) or emergency_queue:
        # Check for emergency
        check_emergencies()

        # If no emergencies, choose the next light
        if not emergency_queue:
            next_light = choose_next_light()
            green_light(next_light)
        
            # Duration could be proportional to queue length
            duration = len(queues[next_light]) * 2
            sleep(duration)  # Allow this light to be green for `duration` seconds, switch to te next queue


    

 # Simulate traffic control
handle_emergencies("east")  # Simulate an emergency in the east direction
adaptive_signal_control()
```
